```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Ensure data is properly loaded
print("Columns in data:", data.columns.tolist())
print("Card Type unique values before cleaning:", data['Card Type'].unique())

# Variables to analyze
analysis_cols = ['HasCrCard', 'Card Type']

# Convert 'Exited' to string (corrected typo from 'Exited')
data['Exited'] = data['Exited'].astype(str)

# Custom colors
custom_colors = ['#E0AFFF', '#FFC0CB']

# Card type standardization
card_type_labels = {
    'GOLD': 'Gold',
    'DIAMOND': 'Diamond',
    'SILVER': 'Silver',
    'PLATINUM': 'Platinum',
    'GOLD CARD': 'Gold',  # Handle common variations
    'DIAMOND CARD': 'Diamond'
}

for col in analysis_cols:
    try:
        # Make copy for analysis
        temp_data = data.copy()
        
        if col == 'Card Type':
            # Convert to string and standardize
            temp_data[col] = temp_data[col].astype(str).str.strip().str.upper()
            temp_data[col] = temp_data[col].map(card_type_labels)
            temp_data = temp_data.dropna(subset=[col])
            
            # Check if we have data left
            if len(temp_data) == 0:
                print(f"No valid Card Type data after cleaning. Skipping...")
                continue
        
        # Create crosstab - ensure we have both Exited values
        ct = pd.crosstab(temp_data[col], temp_data['Exited'], normalize='index')
        
        if ct.empty:
            print(f"No data to plot for {col}. Skipping...")
            continue
            
        # Plot
        plt.figure(figsize=(10, 6))
        ax = ct.plot(kind='bar', stacked=True, color=custom_colors)
        
        # Customizations
        plt.title(f'Churn Rate by {col}', fontsize=14)
        plt.ylabel("Proportion")
        plt.xlabel("Card Type" if col == 'Card Type' else col)
        
        if col == 'Card Type':
            # Ensure consistent order
            ct = ct.reindex(['Gold', 'Diamond', 'Silver', 'Platinum'], errors='ignore')
            ax.set_xticklabels(ct.index, rotation=0)
        
        plt.legend(title="Status", labels=['Stayed', 'Churned'])
        
        # Add percentages
        for p in ax.patches:
            if p.get_height() > 0:
                ax.annotate(
                    f'{p.get_height():.0%}',
                    (p.get_x() + p.get_width()/2, p.get_y() + p.get_height()/2),
                    ha='center', va='center'
                )
        
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"Error processing {col}: {str(e)}")
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from IPython.display import display  # Utile pour un rendu tableau dans Jupyter

# === 1. Cr√©ation du dataset ===
dataa = pd.DataFrame({
    'Card Type': ['DIAMOND', 'DIAMOND', 'DIAMOND', 'GOLD', 'GOLD'],
    'Satisfaction Score': [3, 3, 3, 4, 4],
    'Complain': [1, 0, 0, 0, 0],
    'Exited': [1, 0, 1, 0, 0],
    'Point Earned': [464, 456, 377, 350, 425]
})

# === 2. Transformation des colonnes en cha√Ænes ===
dataa['Satisfaction Score'] = 'Score_' + dataa['Satisfaction Score'].astype(str)
dataa['Exited'] = 'Exited_' + dataa['Exited'].astype(str)
dataa['Complain'] = 'Complain_' + dataa['Complain'].astype(str)

# === 3. Cr√©ation des transactions ===
transactions = dataa[['Card Type', 'Satisfaction Score', 'Exited', 'Complain']]
oht_data = pd.get_dummies(transactions)

# === 4. Application de l'algorithme Apriori ===
frequent_itemsets = apriori(oht_data, min_support=0.3, use_colnames=True)

# === 5. G√©n√©ration des r√®gles d'association ===
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# === 6. Filtrage des r√®gles importantes (lift > 1) ===
rules = rules[rules['lift'] > 1]

# === 7. Formatage du tableau final ===
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].copy()
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(sorted(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(sorted(x)))
rules = rules.sort_values(by='lift', ascending=False).reset_index(drop=True)

# === 8. Affichage du tableau ===
display(rules)
```

# üîç Interpr√©tation des colonnes :
antecedents : l‚Äôensemble de conditions de d√©part.

consequents : ce que l‚Äôon peut d√©duire si les conditions sont vraies.

support : proportion des lignes o√π l‚Äôensemble complet (antecedents + consequents) est pr√©sent.

confidence : probabilit√© que les consequents soient vrais si les antecedents sont vrais.

lift : rapport entre la confiance de la r√®gle et celle qu‚Äôon aurait si antecedent et consequent √©taient ind√©pendants (>1 = d√©pendance positive).

L'analyse des r√®gles d'association r√©v√®le plusieurs relations fortes entre les caract√©ristiques des clients et leur statut de churn (Exited). Les principales observations sont :

Clients fid√®les (Exited_0) :

Une forte association existe entre les clients satisfaits (Score_4), d√©tenteurs d'une carte GOLD et sans r√©clamation (Complain_0). Ces clients ont une probabilit√© de 100% (confidence=1.0) de rester fid√®les, avec un lift √©lev√© (2.5), indiquant une relation significative.

Les clients avec une carte GOLD et un score de satisfaction √©lev√© montrent √©galement une tendance marqu√©e √† ne pas quitter (lift=1.67).

Clients √† risque de churn (Exited_1) :

Les clients ayant une carte DIAMOND et un score de satisfaction moyen (Score_3) sont associ√©s √† un risque accru de churn (lift=1.67). Cela sugg√®re que malgr√© un produit premium, une satisfaction moyenne peut conduire au d√©part.

Les r√©clamations (Complain_0) semblent moins d√©terminantes pour le churn (lift=1.25), mais leur pr√©sence combin√©e √† d'autres facteurs peut influencer la d√©cision.

R√¥le cl√© de la satisfaction :

Un score de satisfaction √©lev√© (Score_4) est syst√©matiquement li√© √† la r√©tention, tandis qu'un score moyen (Score_3) est associ√© au churn. Cela souligne l'importance de maintenir une satisfaction client √©lev√©e.

Implications strat√©giques :

Cibler les d√©tenteurs de carte GOLD avec des avantages suppl√©mentaires pour renforcer leur fid√©lit√©.

Am√©liorer l'exp√©rience des clients DIAMOND, notamment en addressant les causes de satisfaction moyenne.

Surveiller les scores de satisfaction pour identifier pr√©cocement les clients √† risque.



Synth√®se :
La combinaison type de carte, satisfaction et absence de r√©clamations est d√©terminante pour pr√©dire le churn. Une approche cibl√©e sur ces facteurs pourrait r√©duire significativement l'attrition.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

# 1. V√©rifier les colonnes disponibles
print("Colonnes disponibles dans le DataFrame:", df.columns.tolist())

# 2. S√©lectionner uniquement les colonnes num√©riques existantes
colonnes_numeriques = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']
colonnes_disponibles = [col for col in colonnes_numeriques if col in df.columns]

if not colonnes_disponibles:
    raise ValueError("Aucune des colonnes demand√©es n'est pr√©sente dans le DataFrame")

print("Colonnes utilis√©es pour le clustering:", colonnes_disponibles)

# 3. Standardisation des donn√©es
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[colonnes_disponibles])

# 4. Clustering K-means
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# 5. Visualisation des clusters
plt.figure(figsize=(10,6))

# Choisir deux variables pour la visualisation
x_var = 'Age' if 'Age' in df.columns else colonnes_disponibles[0]
y_var = 'Balance' if 'Balance' in df.columns else colonnes_disponibles[1]

# Palette personnalis√©e avec des couleurs rose et mauve
custom_palette =["#E0AFFF", "#FFC0CB", "#B497BD", "#F7C8E0"]   # Rose, Mauve, et une nuance interm√©diaire

# Tracer le graphique avec la palette personnalis√©e
sns.scatterplot(x=df[x_var], y=df[y_var], hue=df['Cluster'], palette=custom_palette)
plt.title(f'Segmentation Client: {x_var} vs {y_var}')
plt.xlabel(x_var)
plt.ylabel(y_var)
plt.show()

# 6. Analyse des clusters
print("\nCaract√©ristiques moyennes par cluster:")
print(df.groupby('Cluster')[colonnes_disponibles].mean())
```

Ce graphique, intitul√© "Segmentation Client: Age vs Balance", illustre la relation entre l'√¢ge des clients et le solde de leur compte. On observe que la distribution des soldes est relativement dispers√©e, avec des valeurs allant de 0 √† 150 000 unit√©s mon√©taires. Les clients plus jeunes (20-30 ans) tendent √† avoir des soldes plus faibles, tandis que les tranches d'√¢ge interm√©diaires (30-60 ans) pr√©sentent une plus grande variabilit√©, avec certains individus affichant des soldes √©lev√©s. En revanche, les clients plus √¢g√©s (70-90 ans) montrent une tendance √† des soldes plus mod√©r√©s, sans atteindre les niveaux maximaux observ√©s chez les tranches d'√¢ge pr√©c√©dentes. Cela pourrait sugg√©rer que les clients d'√¢ge moyen sont plus actifs financi√®rement ou accumulent davantage de fonds, tandis que les jeunes et les seniors ont des profils de balance distincts, influenc√©s par des facteurs tels que l'√©pargne, les d√©penses ou les besoins financiers sp√©cifiques √† leur √©tape de vie.

üìä Conclusion de la segmentation K-means :
La segmentation K-Means a permis de regrouper les clients en trois clusters distincts selon leur √¢ge, balance, revenu estim√© et score de cr√©dit.
L‚Äôanalyse visuelle montre que :

Cluster 2 contient majoritairement des clients avec une balance faible voire nulle, quel que soit l‚Äô√¢ge.

Clusters 0 et 1 regroupent des clients avec des balances plus √©lev√©es, ce qui refl√®te un potentiel financier plus int√©ressant.

```python
import pandas as pd

# Scikit-Learn imports
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    roc_curve,
    confusion_matrix
)

# For plotting
import matplotlib.pyplot as plt
import seaborn as sns

# ‚îÄ‚îÄ‚îÄ 1) V√©rification des colonnes ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
required = ['Exited', 'Cluster']
missing = [c for c in required if c not in df.columns]
if missing:
    raise ValueError(f"Colonnes manquantes dans df : {missing}")

# ‚îÄ‚îÄ‚îÄ 2) S√©paration target / features ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
y = df['Exited'].astype(int)                # cible binaire
X = df.drop(['Exited', 'Cluster'], axis=1)  # features

# ‚îÄ‚îÄ‚îÄ 3) Typage des variables ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
numeric_features     = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object', 'category']).columns.tolist()

# ‚îÄ‚îÄ‚îÄ 4) Pr√©processeur ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(drop='first'), categorical_features),
])

# ‚îÄ‚îÄ‚îÄ 5) D√©finition des pipelines ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
models = {
    'LogisticRegression': Pipeline([
        ('preproc', preprocessor),
        ('clf', LogisticRegression(max_iter=1000))
    ]),
    'RandomForest': Pipeline([
        ('preproc', preprocessor),
        ('clf', RandomForestClassifier(n_estimators=100))
    ]),
    'SVM': Pipeline([
        ('preproc', preprocessor),
        ('clf', SVC(probability=True))
    ]),
}

# ‚îÄ‚îÄ‚îÄ 6) S√©paration hold-out ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# 10% en test final, 90% pour CV
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.10,
    stratify=y,
    random_state=42
)

# ‚îÄ‚îÄ‚îÄ 7) Validation crois√©e 5-fold sur X_train ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
cv      = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']

print("=== R√©sultats de la CV 5-fold ===")
roc_auc_means = {}

for name, pipeline in models.items():
    cv_res = cross_validate(
        pipeline,
        X_train, y_train,
        cv=cv,
        scoring=scoring,
        return_train_score=False,
        n_jobs=-1
    )
    print(f"\n>> {name}")
    for metric in scoring:
        mean = cv_res[f'test_{metric}'].mean()
        std  = cv_res[f'test_{metric}'].std()
        print(f"   ‚Ä¢ {metric:10s} : {mean:.3f} ¬± {std:.3f}")
    roc_auc_means[name] = cv_res['test_roc_auc'].mean()

# ‚îÄ‚îÄ‚îÄ 8) S√©lection du meilleur mod√®le ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
best_name     = max(roc_auc_means, key=roc_auc_means.get)
best_pipeline = models[best_name]
print(f"\n>>> Meilleur mod√®le (CV AUC) : {best_name} = {roc_auc_means[best_name]:.3f}")

# ‚îÄ‚îÄ‚îÄ 9) Entra√Ænement de tous les pipelines sur X_train ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
for pipeline in models.values():
    pipeline.fit(X_train, y_train)

# ‚îÄ‚îÄ‚îÄ 10) √âvaluation finale + visualisations ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

# 10.1 Courbes ROC compar√©es
plt.figure(figsize=(10, 8))
for name, pipeline in models.items():
    y_proba = pipeline.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc = roc_auc_score(y_test, y_proba)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], '--', label='Al√©atoire')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Courbes ROC compar√©es')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()

# 10.2 Matrices de confusion en heatmaps
for name, pipeline in models.items():
    y_pred = pipeline.predict(X_test)
    cm     = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Matrice de confusion ‚Äî {name}')
    plt.xlabel('Pr√©dit')
    plt.ylabel('R√©el')
    plt.tight_layout()
    plt.show()
```

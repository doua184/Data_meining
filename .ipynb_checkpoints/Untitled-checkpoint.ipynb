Le dataset Supermarket Sales contient des donn√©es sur les transactions effectu√©es dans un supermarch√© fictif. Il permet d‚Äôanalyser les comportements d‚Äôachat des clients selon plusieurs dimensions : lieu, genre, m√©thode de paiement, type de produit, etc.

```python
# Importation des librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
print(df.head())
```

üìä Description du graphique ‚Äì Modes de paiement utilis√©s
Le graphique pr√©sente la fr√©quence d‚Äôutilisation des diff√©rents modes de paiement par les clients du supermarch√©. Il s‚Äôagit d‚Äôun diagramme en barres verticales o√π chaque barre repr√©sente un mode de paiement.
Les trois modes de paiement disponibles sont :
üí≥ Credit card /üíµ Cash/üì± Ewallet
Les hauteurs des barres indiquent le nombre de transactions effectu√©es avec chaque mode.

Les barres sont g√©n√©ralement assez √©quilibr√©es, ce qui montre que les clients utilisent tous les moyens de paiement de mani√®re assez homog√®ne, avec de l√©g√®res variations selon les pr√©f√©rences.

Si la barre pour Ewallet est la plus haute, cela signifie que les portefeuilles √©lectroniques sont le mode de paiement le plus utilis√© dans ce supermarch√©.

```python
sns.countplot(data=df, x="Payment")
plt.title("Modes de paiement utilis√©s")
plt.show()
```

üì¶ Description du graphique ‚Äì Total des ventes par genre
Le graphique affiche un boxplot (ou bo√Æte √† moustaches) qui compare la distribution du montant total des ventes (Total) entre les deux genres : Male et Female.

üîπ Lecture du boxplot :
Chaque bo√Æte repr√©sente la distribution des montants d√©pens√©s par genre.

La ligne au centre de chaque bo√Æte indique la m√©diane (le montant au milieu de la distribution).

Les bords de la bo√Æte montrent le 1er et 3e quartile (50% des donn√©es).

Les "moustaches" repr√©sentent l‚Äô√©tendue des donn√©es, et les points isol√©s sont les valeurs extr√™mes (outliers).

```python
sns.boxplot(data=df, x="Gender", y="Total")
plt.title("Total des ventes par genre")
plt.show()
```

üì¶ Boxplot : Prix unitaire par ligne de produit
Un boxplot (ou bo√Æte √† moustaches) permet de visualiser la distribution d'une variable (ici, Unit price) pour chaque groupe (ici, Product line).

‚úÖ Ce que le graphique montre :
La bo√Æte repr√©sente les 50 % centraux des donn√©es (entre le 1er quartile Q1 et le 3e quartile Q3).

La ligne √† l‚Äôint√©rieur de la bo√Æte est la m√©diane (Q2).

Les moustaches montrent les valeurs minimum/maximum sans les outliers.

Les points isol√©s sont des valeurs extr√™mes (outliers).

```python
sns.boxplot(x='Product line', y='Unit price', data=df)
plt.title("Prix unitaire par ligne de produit")
plt.xticks(rotation=45)
plt.show()
```

# <font color=red> Histogramme

# <font color=PINK> PRICE UNITAIRE 
Analyse : Les prix unitaires sont r√©partis de mani√®re assez uniforme entre 10 et 100.

Interpr√©tation : Il n'y a pas de concentration de prix sp√©cifique, ce qui sugg√®re une diversit√© de produits avec une large gamme de prix.

```python
sns.histplot(df["Unit price"], kde=True, bins=30, color='lightblue')
plt.title("Histogramme des prix unitaires (Unit price)")
plt.xlabel("Prix unitaire")
plt.ylabel("Fr√©quence")
plt.show()
```

# üìà Courbe de densit√© de la variable Quantity

Elle repr√©sente la distribution "liss√©e" du nombre de produits achet√©s par les clients.

Contrairement √† un histogramme en barres, la courbe de densit√© trace une courbe continue qui montre o√π les valeurs sont les plus fr√©quentes.

Elle permet de voir plus facilement la forme globale de la distribution.

```python
sns.kdeplot(df["Quantity"], fill=True)
plt.title("Densit√© de la quantit√© achet√©e")
plt.xlabel("Quantit√©")
plt.ylabel("Densit√©")
plt.show()
```

Ce code g√©n√®re un histogramme d'une colonne d'un DataFrame (probablement appel√© df) nomm√©e "Tax 5%", avec une courbe d'Estimation de Densit√© par Noyau (KDE) superpos√©e. 

Si le pic de la courbe KDE est situ√© vers une valeur faible : cela pourrait indiquer que la majorit√© des taxes appliqu√©es sont faibles, peut-√™tre en raison d‚Äôune faible imposition ou d‚Äôun taux d'imposition r√©duit.

Si l'histogramme est tr√®s large et l√©g√®rement √©tendu vers des valeurs plus √©lev√©es : cela pourrait signifier qu'il y a une certaine variation dans le montant des taxes, avec quelques valeurs beaucoup plus √©lev√©es que la moyenne.

Si l'histogramme montre un pic avec peu de variation : cela indiquerait que la plupart des taxes sont concentr√©es autour d'une valeur sp√©cifique, avec peu de fluctuations.

```python
sns.histplot(df["Tax 5%"], kde=True, bins=30, color='salmon')
plt.title("Histogramme des taxes (5%)")
plt.xlabel("Montant de la taxe")
plt.ylabel("Fr√©quence")
plt.show()
```
